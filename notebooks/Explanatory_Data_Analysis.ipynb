```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../data/raw/Mall_Customers.csv")
df.head()
```

```python
print(df.shape)
print(df.describe)
print(df.info())
print(df.dtypes)
```

```python

df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
plt.figure(figsize=(8,6))
sns.histplot(x=df["Spending Score (1-100)"],bins=20,kde=True)
plt.title("distribution of Spending score")
plt.xlabel("Spending Score (1-100)")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df["Genre"],y=df["Spending Score (1-100)"],data=df)
plt.title("distribution of Spending score")
plt.xlabel("Spending Score (1-100)")
plt.ylabel("count")
plt.show()
```

```python
cols=["Spending Score (1-100)","Annual Income (k$)","Age"]
corr=df[cols].corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("correletion matrix")
plt.show()
```
